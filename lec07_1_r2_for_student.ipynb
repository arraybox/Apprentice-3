# R2

```python
#@title { vertical-output: true}
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(44)

# data set
x = np.array([1.78, 6.12, 6.63, 4.13, 8.55, 10.92, 4.81, 3.75, 2.28, 5.74]).reshape(-1, 1)
y = np.array([1.06, 3.04, 3.93, 2.19, 4.89, 5.37, 2.91, 2.36, 1.13, 3.05])
plt.plot(x, y, '.')
plt.show()
```

```python
#@title { vertical-output: true}
from sklearn.linear_model import LinearRegression

model_linear = LinearRegression()
model_linear.fit(x, y)

x_test = np.linspace(0, 10, 100).reshape(100, 1)
y_test = model_linear.predict(x_test)
plt.plot(x, y, '.')
plt.plot(x_test, y_test)
plt.show()
```

```python
#@title { vertical-output: true}
from sklearn.metrics import r2_score

def my_r2_score(y_true, y_pred):

    # 1. 실제값(y_true)의 평균을 계산합니다.
    y_true_mean = np.mean(y_true)
    
    # 2. 총 변동량(Total Sum of Squares, SST)을 계산합니다.
    # (실제값 - 실제값의 평균)의 제곱의 합
    ss_tot = np.sum((y_true - y_true_mean)**2)
    
    # 3. 잔차의 제곱합(Sum of Squares of Residuals, SSR)을 계산합니다.
    # (실제값 - 예측값)의 제곱의 합
    ss_res = np.sum((y_true - y_pred)**2)
    
    # 4. R2 스코어를 계산합니다: 1 - (SSR / SST)
    r2 = 1 - (ss_res / ss_tot)

y_hat = model_linear.predict(x)

r2 = my_r2_score(y.reshape(-1, 1), y_hat.reshape(-1, 1))
print(f'R2: {r2}')

r2 = r2_score(y.reshape(-1, 1), y_hat.reshape(-1, 1))
print(f'R2: {r2}')
```
